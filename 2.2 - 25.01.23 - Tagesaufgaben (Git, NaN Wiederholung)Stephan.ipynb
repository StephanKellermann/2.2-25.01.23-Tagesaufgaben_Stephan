```python
import pandas as pd
import missingno as msno
```

# Step1
### Lese die  'Dataset.csv' Datei  und speichere den Dataframe in der Variable 'data'. Gib anschließend die ersten 4 Zeilen des Dataframes aus.


```python
df = pd.read_csv("Dataset.csv")
df.head(10)
```

# Step2
## Berechne die Menge an NaNs in jeder Spalte und lass dir die 10 Spalten mit den meisten NaNs anzeigen.


```python
nans_per_column = df.isna().sum()
nans_per_column = nans_per_column.sort_values(ascending=False)
nans_per_column.head(11)
```

# Step 3
### Lösche die Spalten in denen mehr als 50% der Werte NaNs sind. Überschreibe dazu deinen DataFrame.

```python
# Berechne die Prozentzahl der NaN-Werte pro Spalte
nan_percentage_per_column = df.isna().mean()
# Entferne die Spalten, bei denen mehr als 50% der Werte NaNs sind
df = df.drop(columns=nan_percentage_per_column[nan_percentage_per_column > 0.5].index)
df.head(10)
```

# Step 4 - Optional
### Prozentual gesehen, wie viele NaN Werte gibt es im ganzen DataFrame?

```python
# Berechne die Anzahl der NaNs im DataFrame insgesamt
total_nans = df.isna().sum().sum()

# Berechne die Gesamtzahl der Werte im DataFrame
total_values = df.size

# Berechne die Prozentzahl der NaN-Werte
nan_percentage = (total_nans / total_values) * 100

print("There are " + str(nan_percentage) + "% NaN-Werte.")
```

# Step 5
### Zeige mithilfe der missingno Library alle NaN Werte in einem Graphen an

```python
# Zeige den Graph der fehlenden Werte an
msno.matrix(df)
```

# Step 7
## Ersetze die NaN Werte in der Spalte "Number of Existing Stories"  mit einer 1

```python
# Ersetze die NaN-Werte in der Spalte "Number of Existing Stories" mit einer 1
df["Number of Existing Stories"].fillna(1) # erst sagt man wo man hingeht und dann was man macht # inplace=True geht (aber druckt dann nicht aus)
```

# Step 8
## Ersetze die NaN Werte in der Spalte "Existing Units"  mit dem Maximalwert dieser Spalte

```python
df["Existing Units"].fillna(df["Existing Units"].max())
```
